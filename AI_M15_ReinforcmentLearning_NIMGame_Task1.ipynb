# Assigment 1 - (Me "Computer" First Start The Game)

```python
import numpy as np
import random

A = np.array([0, 0.5])
print("A = ", A)
B = np.array([0, 0.5, 0.5, 0.5])
print("B = ", B)
C = np.array([0, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5])
print("C = ", C)
D = np.array([0, 0.5, 0.5, 0.5, 0.5, 1, 0.5, 0.5])
print("D = ", D)
E = np.array([0, 0.5, 0, 0, 0, 0, 0])
print("E = ", E)
F = np.array([0.0, 1.0])
print("F = ", F)
print()

# ME => "COMPUTER"
# ENEMY => ENEMY
#===================== ME -VS- ENEMY -VS- ME -VS- ENEMY -VS- ME

g = 0
j = 0
k = 0
L = 0
m = 0
n = 0

Iteration = 100
alpha = 0.1
Exploration_rate = 0.8
P = 0

for i in range(Iteration):
    # ================================ A Proses INITIALIZATION
    g = 1
    
    # ================================ B-j Proses COMPUTER
    # ========================================================== COMPUTER PROSES
    P = random.random()
    if(P < Exploration_rate):
        j = random.randint(1,3)
    else:
        # >>>>>>>>>>>>>>>>>>>>> MAKMSIMUM
        if(B[1] >= B[2] and B[1] >= B[3]):
            j = 1
        elif(B[2] >= B[1] and B[2] >= B[3]):
            j = 2
        elif(B[3] >= B[1] and B[3] >= B[2]):
            j = 3
        # >>>>>>>>>>>>>>>>>>>>> MAKMSIMUM    
    A[g] = A[g] + alpha*(B[j] - A[g])
    # ========================================================== COMPUTER PROSES
    
    if(j == 1):
        # ================================ C-k Proses ENEMY
        k = random.randint(1,2)
        B[j] = B[j] + alpha*(C[k] - B[j])
        
        if(k == 1):
            # ================================ D-L Proses COMPUTER
            # ========================================================== COMPUTER PROSES
            P = random.random()
            if(P < Exploration_rate):
                L = random.randint(1,2)
            else:
                # >>>>>>>>>>>>>>>>>>>>> MAKMSIMUM
                if(D[1] >= D[2]):
                    L = 1
                else:
                    L = 2
                # >>>>>>>>>>>>>>>>>>>>> MAKMSIMUM    
            C[k] = C[k] + alpha*(D[L] - C[k])
            # ========================================================== COMPUTER PROSES
            
            if(L == 1):
                #print("==> WIN SMART <==")
                # ================================ E-m Proses ENEMY
                m = 1
                D[L] = D[L] + alpha*(E[m] - D[L])
                
                # ================================ F-n Proses COMPUTER
                n = 1
                E[m] = E[m] + alpha*(F[n] - E[m])
                
            elif(L == 2):
                #print("==> IMPOSIBLE <==")
                # ================================ E-m Proses ENEMY
                m = 2
                D[L] = D[L] + alpha*(E[m] - D[L])
                
        elif(k == 2):
            #print("==> LOSE by Smart ENEMY <==")
            # ================================ D-L Proses COMPUTER
            L = 3
            C[k] = C[k] + alpha*(D[L] - C[k])
            
            # ================================ E-m Proses ENEMY
            m = 3
            D[L] = D[L] + alpha*(E[m] - D[L])
                
    elif(j == 2):
        # ================================ C-k Proses ENEMY
        k = random.randint(3,4)
        B[j] = B[j] + alpha*(C[k] - B[j])
        
        if(k == 3):
            #print("==> LOSE by Smart ENEMY <==")
            # ================================ D-L Proses COMPUTER
            L = 4
            C[k] = C[k] + alpha*(D[L] - C[k])
            
            # ================================ E-m Proses ENEMY
            m = 4
            D[L] = D[L] + alpha*(E[m] - D[L])
            
        elif(k == 4):
            # ================================ D-L Proses COMPUTER
            L = 5
            C[k] = C[k] + alpha*(D[L] - C[k])
            #print("==> WIN By Stupid ENEMY <==")
            
    elif(j == 3):
        #print("==> IMPOSIBLE <==")
        # ================================ C-k Proses ENEMY
        k = random.randint(5,6)
        B[j] = B[j] + alpha*(C[k] - B[j])
        
        if(k == 5):
            # ================================ D-L Proses COMPUTER
            L = 6
            C[k] = C[k] + alpha*(D[L] - C[k])
            
            # ================================ E-m Proses ENEMY
            m = 5
            D[L] = D[L] + alpha*(E[m] - D[L])
            
        elif(k == 6):
            # ================================ D-L Proses COMPUTER            
            L = 7
            C[k] = C[k] + alpha*(D[L] - C[k])
        
            # ================================ E-m Proses ENEMY
            m = 6
            D[L] = D[L] + alpha*(E[m] - D[L])
            
    
    print(i+1, "Step-by-step =>",g,j,k,L,m,n)
    if((i+1) % 1 == 0):
        print("A = ", A)
        print("B = ", B)
        print("C = ", C)
        print("D = ", D)
        print("E = ", E)
        print("F = ", F)
    print()
    g = 0
    j = 0
    k = 0
    L = 0
    m = 0
    n = 0
```
